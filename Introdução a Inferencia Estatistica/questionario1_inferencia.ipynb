```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def poder_teste(mu0, mu1, sigma, alpha, n):
  valor_critico = stats.norm.ppf(1 - alpha/2)
  erro_padrao_amostral = sigma / np.sqrt(n)
  regiao_rejeicao = valor_critico * erro_padrao_amostral
  beta = stats.norm.cdf(regiao_rejeicao, loc=mu1, scale=erro_padrao_amostral)
  poder = 1 - beta
  return poder
```

Variando alpha

```python
mu0 = 0 # H0: hipótese nula
mu1 = 1 # H1: hipótese alternativa - um dos valores possiveis para H1 (mu != 0)
sigma = 1 # desvio padrão
valores_alpha = np.arange(0.01, 0.11, 0.01) # nível de significância variando
n = 30 # tamanho da amostra

# cálculo do poder com mu1 e n fixos
poder_valores = []
for alpha in valores_alpha:
  poder = poder_teste(mu0, mu1, sigma, alpha, n)
  poder_valores.append(poder)

plt.figure(figsize=(10, 6))
plt.plot(valores_alpha, poder_valores, label='n = {}, mu1 = {}'.format(n, mu1))
plt.xlabel('Alpha')
plt.ylabel('Poder')
plt.title('Função poder para teste de hipóteses com n e mu1 fixos')
plt.legend()
plt.show()
```

Variando n

```python
mu0 = 0 # H0: hipótese nula
mu1 = 1 # H1: hipótese alternativa - um dos valores possiveis para H1 (mu != 0)
sigma = 1 # desvio padrão
alpha = 0.05 # nível de significância variando
valores_n = range(5, 41, 5) # tamanho da amostra

# cálculo do poder com mu1 e n fixos
poder_valores = []
for n in valores_n:
  poder = poder_teste(mu0, mu1, sigma, alpha, n)
  poder_valores.append(poder)

plt.figure(figsize=(10, 6))
plt.plot(valores_n, poder_valores, label='alpha = {}, mu1 = {}'.format(alpha, mu1))
plt.xlabel('n (Tamanho da amostra)')
plt.ylabel('Poder')
plt.title('Função poder para teste de hipóteses com alpha e mu1 fixos')
plt.legend()
plt.show()
```

Variando mu1 (H1)

```python
mu0 = 0 # H0: hipótese nula
valores_mu1 = np.round(np.linspace(1, 2, 10), 2) # H1: hipótese alternativa - alguns valores possiveis para H1 (mu != 0)
sigma = 1 # desvio padrão
alpha = 0.05 # nível de significância variando
n = 30 # tamanho da amostra

# cálculo do poder com mu1 e n fixos
poder_valores = []
for mu1 in valores_mu1:
  poder = poder_teste(mu0, mu1, sigma, alpha, n)
  poder_valores.append(poder)

plt.figure(figsize=(10, 6))
plt.plot(valores_mu1, poder_valores, label='alpha = {}, n = {}'.format(alpha, n))
plt.xlabel('Hipótese Alternativa (H1)')
plt.ylabel('Poder')
plt.title('Função poder para teste de hipóteses com alpha e n fixos')
plt.legend()
plt.show()
```
